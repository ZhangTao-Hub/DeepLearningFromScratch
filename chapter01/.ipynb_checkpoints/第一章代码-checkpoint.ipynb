# 类

```python
class Man():
    def __init__(self, name):
        self.name = name
        print("Initialized!")

    def hello(self):
        print("Hello" + self.name + "!")

    def goodbye(self):
        print("Hello" + self.name + "!")
```

```python
man = Man("David")
```

```python
man.hello()
```

```python
man.goodbye()
```

# Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.arange(0, 6, 0.1)
```

```python
y = np.sin(x)
```

```python
plt.plot(x, y)
```

```python
x = np.arange(0, 6, 0.1)
```

```python
y1 = np.sin(x)
```

```python
y2 = np.cos(x)
```

```python
plt.plot(x, y1, label="sin")
plt.plot(x, y2, linestyle='--', label='cos')
plt.xlabel('x')
plt.ylabel('y')
plt.title('sin & cos')
plt.legend()
```
